```python
# import the libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Basic of the img i.e accessing and displaying...
# import the img

my_img = cv2.imread('G:\\images\\Davis_1.jpg') # reading img
cv2.imshow('image',my_img) # displaying the img
cv2.waitKey(0) # until you press any key in your keyboard then destory the img or vedio etc.
cv2.destroyAllWindows()
```

```python
face_cascade = cv2.CascadeClassifier('C:\\Users\\pavan tanniru\\Downloads\\opencv\\frontalface_default.xml')
gray = cv2.cvtColor(my_img, cv2.COLOR_BGR2GRAY)

# Our classifier returns the ROI of the detected face as a tuple
# It stores the top left coordinate and the bottom right coordiantes
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

# When no faces detected, face_classifier returns and empty tuple
if faces is ():
    print("No faces found")

# We iterate through our faces array and draw a rectangle
# over each face in faces
for (x,y,w,h) in faces:
    cv2.rectangle(my_img, (x,y), (x+w,y+h), (127,0,255), 2)
    cv2.imshow('Face Detection', my_img)
    cv2.waitKey(0)
    
cv2.destroyAllWindows()
```
